# 파이썬 기초 문법
### 1. 숫자형

```python
a = 123
print(a)

a = -178
print(a)
```

```python
a = 1.2
print(a)

a = -3.45
print(a)
```

```python
# 4.24 * 10의 -10
a = 4.24e-10
print(a)
```

```python
# 8진수 : 0o
# 16진수 : 0x

a = 0o177
print(a)

a = 0x8ff
print(a)
```

```python
# + , - , * , / , // , % , **

a = 3
b = 4

print(a + b)
print(a - b)
print(a * b)
print(a / b) # 나눗셈
print(a // b) # 몫
print(a % b) # 나머지
print(a ** b) # 제곱
```
