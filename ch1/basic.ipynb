```python
# https://naver.com

# 비밀먼호 생성
# 규칙 1. https:// 제외 => naver.com
# 규칙 2. 처음 만나는 . 이후 부분은 제외 => naver
# 규칙 3. 남은 글자 중 처음 세자리 + 글자 갯수 + 글자 내 'e' 개수 + '!' 비밀번호 생성 => nav51!

# 문자 함수, 슬라이싱, str()

address = "https://naver.com"
address = address.replace("https://","")
pos=address.find(".")
address = address[:pos]
password = address[:3] + str(len(address)) + str(address.count("e")) + "!"
print("비밀번호는", password, "입니다.")
```
