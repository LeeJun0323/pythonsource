```python
# random : 난수(규칙이 없는 임의의 수)를 발생시키는 모듈

import random

# random() : 0.0 ~ 1.0 사이의 실수
# 0.0 <= random.random() < 1.0
print(random.random())
# 시작숫자 <= random.random() <= 끝숫자
print(random.randint(1,10))
# uniform(최소, 최대) : 지정한 두 실수 사이의 난수
print(random.uniform(10,20))
# randrange(stop)
# randrange(start, stop)
# randrange(start, stop, step)
print(random.randrange(10))
# choice(시퀀스) : 리스트. 튜플, 문자열
print(random.choice([1,2,3,4,5,6,7]))
# shuffle()
list1 = list(range(1,8))
random.shuffle(list1)
print(list1)

# sample(시퀀스, 개수) : 리스트, 튜플, 문자열
print(random.sample(list1, k = 2))
print(random.sample("Python", k = 2))
```

```python
# 예외 클래스 생성 가능

class MyError(Exception) :
    def __str__(self) :
        return "허용되지 않는 별명입니다."

def say_nick(nick) :
    if nick == "바보" :
        raise MyError()
    print(nick)


try :
    say_nick("천사")
    say_nick("바보")
except MyError as e :
    print(e)
```

```python
# 문제 1. 랜덤 비밀번호 생성기

# 사용자로부터 생성할 비밀번호의 길이를 입력받아 랜덤 비밀번호를 생성하는 프로그램 작성
# choice()
chars = "ABCDEFGHIJKLMNOPQRSTUVWXZ0123456789!@#$%"
# 비밀번호 길이는 4 이상만 허용
# 숫자가 아닌 값을 입력하면 예외 처리한다.

import random

def makePassword():
    while True :
        try :
            password = int(input("생성할 비밀번호의 길이를 입력하세요."))
            return password
        except :
            print("숫자가 아닌 값을 입력하였습니다.")
            continue
    
user_password = ""

while True :
    pd = makePassword()
    
    if pd < 4 :
        print("4자리 이상이여야 합니다.")
    else :
        for i in range(pd) :
            user_password += random.choice(chars)
        break         
print(f'랜덤 비밀번호 : {user_password}')
```

```python
# 나이 계산 프로그램
# 생년월일을 입력받아 현재 나이를 계산하는 프로그램 작성
# 입력형식 : 2001-08-25
# input() => "2001-08-25" => 날짜 형식
# datetime.strptime("2001-08-25",%Y-%m-%d)

import datetime

# birth = input("생년월일을 입력하세요(ex 2001-08-25) : ")

# birth = str(datetime.datetime.strptime(birth, "%Y-%m-%d")).split("-")
# day = str(datetime.datetime.now()).split("-")
# Mon = int(day[0]) - int(birth[0]) + 1
# print(f'{Mon}살')

# 오늘날짜
now = datetime.datetime.now()
try :
    # 생년월일 입력
    birth_str = input("생년월일 입력(YYYY-MM-DD)")
    # '2001-05-15'
    birth = datetime.datetime.strptime(birth_str, "%Y-%m-%d")

    # 2026 - 2001
    # now 년도만 추출
    age = now.year - birth.year

    if (now.month, now.day) < (birth.month, birth.day) :
        age -= 1
    print(f'나이 : {age}')
except :
    print("날짜 형식이 올바르지 않습니다.")
```

```python
# 다음과 같은 csv 파일이 있다.
# 이름, 국어, 영어, 수학
# 홍길동, 90, 80, 70
# 김철수, 90, 80, 70
# 이영희, 90, 80, 70
# 성춘향, 90, 80, 70

# csv 파일을 읽어 학생별 평균 점수 출력
# 홍길동 : 80.0

# csv.DictReader() 사용, 평균은 둘째자리까지

import csv

with open("./data/student.csv", "w", encoding="utf-8", newline = "") as f :
    headers = ['이름', '국어', '영어', '수학']
    writer = csv.DictWriter(f, fieldnames=headers)
    # 헤더작성
    writer.writeheader()
    writer.writerow({"이름" : "홍길동", "국어" : 88, "영어" : 92, "수학" : 70})
    writer.writerow({"이름" : "김철수", "국어" : 85, "영어" : 70, "수학" : 80})
    writer.writerow({"이름" : "이영희", "국어" : 92, "영어" : 88, "수학" : 76})
    writer.writerow({"이름" : "성춘향", "국어" : 93, "영어" : 82, "수학" : 77})


with open("./data/student.csv", "r", encoding="utf-8") as f :
    reader = csv.DictReader(f)
    for row in reader :
        avg = (int(row.get('국어')) + int(row.get('영어')) + int(row.get('수학'))) / 3
        print(f'{row.get('이름')} : {avg:.2f}')
```
