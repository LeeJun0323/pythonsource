````python
# add
# add(2,3)
# add(2,3,4,5)
# add(2,3,4,5,6,7)

# 가변 매개변수
# 입력값이 몇 개가 될 지 모르는 경우
# * 
# def 함수명(*변수명) :
#     해야할 일

def add(*args) :
    result = 0
    for i in args :
        result = result + i
    print(result)
````

````python
# def add_mul(choice, *args, a) :
#     pass

# 일반 매개변수는 앞 쪽으로 선언 
# +, * 실행 
def add_mul(choice, *args) :
    if choice == "add" :
        result = 0
        for i in args :
            result = result + i
    elif choice == "mul" :
        result = 1
        for i in args :
            result = result * i

    return result
````

````python
# add_mul("add", 10, 11, 12, 13)
print(add_mul("add", 10, 11, 12, 13))
print(add_mul("mul", 10, 11, 12, 13))
````

````python
# 키워드 매개변수 (key = vlaue)
# **변수명

# 딕셔너리 == 키워드 매개변수
def print_kwargs(**kwargs) :
    print(kwargs)

# key 값만 출력
    for k in kwargs.keys() :
        print(k)
# value 값만 출력
    for v in kwargs.values() :
        print(v)
# key, value 같이 출력
    for k,v in kwargs.items() :
        print(k,v)
````

````python
print_kwargs(name = "kim", age = 15)
print_kwargs(name = "park", age = 25, addr = "서울")
print_kwargs(name = "hong", age = 30, addr = "경기", tel = "010-1234-5678")
````

````python
# 변수의 효력 범위

# 함수 밖의 변수 : 전역변수
a = 1
def vertest(a) :
    # 매개변수 a : 지역변수
    a = a + 1

vertest(a)
print(a)
````

````python
# 전역변수 a 값 변경하기
# 1) 리턴 이용 (추천방법)

a = 1
def vertest(a) :
    # 매개변수 a : 지역변수
    a = a + 1
    return a

a = vertest(a)
print(a)
````

````python
# 2) global 키워드 사용
# 함수 안에서 전역변수 접근 안됨
# UnboundLocalError: cannot access local variable 'a' where it is not associated with a value

a = 1
def vertest() :
    global a
    a = a + 1

vertest()
print(a)
````

````python
# 사칙연산 함수
# 결과 리턴

def calculator(a, b, sign) :
    if sign == "+" :
        return a + b
    elif sign == "-" :
        return a - b
    elif sign == "*" :
        return a * b
    elif sign == "/" :
        return a / b
    elif sign == "//" :
        return a // b
    else :
        return
    
print(calculator(10, 20, "+"))
print(calculator(10, 20, "-"))
print(calculator(10, 20, "*"))
print(calculator(10, 20, "/"))
print(calculator(10, 20, "//"))
````

````python
# filter()
list1 = list(range(1,21))

# filter(함수, iterable)

# 짝수만 새로운 리스트로 생성
# 방법1
even_list = []
for i in list1 :
    if i % 2 == 0 :
        even_list.append(i)
print(even_list)

# 방법2 (리스트 컴프리헨션)
print([i for i in list1 if i % 2 == 0])

# 방법3 filter
# filter(함수, iterable)
print(list(filter(lambda i : i % 2 ==0, list1)))
````

````python
strings = ["Bob", "charles", "Alexander3", "teddy", "apple", "Banana"]

# 단어의 시작이 대문자인 단어들만 모아서 새로운 리스트로 생성
upper_strings = []
for i in strings :
    if i[0].isupper() :
        upper_strings.append(i)

print(upper_strings)

# 리스트 컴프리헨션
print([i for i in strings if i[0].isupper() ])

# filter(함수, 원본리스트) : 새로운 리스트로 생성
print(list(filter(lambda x : x[0].isupper(), strings)))
````

````python
def calculator(num1, num2, op) :
    if op == "+" :
        return num1 + num2
    elif op == "-" :
        return num1 - num2
    elif op == "*" :
        return num1 * num2
    elif op == "/" :
        return num1 / num2
    elif op == "//" :
        return num1 // num2
    else :
        return

# 숫자 2개 입력받기, 연산자 기호(+,-,*....) 입력받기

num1 = int(input("첫번째 숫자를 입력해주세요."))
num2 = int(input("두번째 숫자를 입력해주세요."))
op = input("연산 기호(+, -, *, /, //)를 입력해주세요.")
print(calculator(num1,num2,op))
````
