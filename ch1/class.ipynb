| `@staticmethod` | 없음 | 없음(매개변수로 받은 것만) | 클래스와 관련은 있지만 독집적인 계산 |

```python
# Account 클래스 작성
# 멤버변수 : 계좌주(owner), 잔액(balance)
# 멤버 메소드 : desposit(money) : 잔액 = 잔액 + 입금액
# 멤버 메소드 : withdraw(money) : 잔액 = 잔액 - 출금액
# 멤버 메소드 : showBalance(money) : 현재 잔액 출력

class Account :
    
    def __init__(self, owner, balance):
        self.owner = owner
        self.balance = balance

    def desposit(self,money) :
        self.balance += money
    
    def withdraw(self,money) :
        if self.balance < money :
            print("잔액부족")
        else :    
            self.balance -= money

    def showBalance(self) :
        print(f'현재 {self.owner}님의 잔액은 {self.balance}원 남았습니다.')

mon = Account("홍길동", 10000)
mon.showBalance()
mon.desposit(20000)
mon.showBalance()
mon.withdraw(5000)
mon.showBalance()
```

```python
# Order 클래스 작성
# 클래스 변수 : total_order
# 멤버 변수 : 메뉴(menu), 가격(price), 수량(count)
# 멤버 메소드 : getTotlaPrice() = 가격 x 수량 : 총 주문금액 반환
# 멤버 메소드 : showOrder() = 출력문 작성
#                   메뉴 : 아메리카노
#                   가격 : 4000원
#                   수량 : 2개
#                   총 주문금액 : 8000원

class Order :
    
    total_order = 0

    def __init__(self, menu, price, count):
        self.menu = menu
        self.price = price
        self.count = count
        Order.total_order += 1
    
    def getTotalPrice(self) :
        return self.price * self.count

    def showOrder(self) :
        print(f'메뉴 : {self.menu}')
        print(f'가격 : {self.price}')
        print(f'수량 : {self.count}')
        print(f'총 주문금액 : {self.getTotalPrice()}')
        

order1 = Order("아메리카노", 4000, 2)
order2 = Order("카푸치노", 4000, 2)

order1.showOrder()
print()
order2.showOrder()
print()
print(f'전체 주문 수 : {Order.total_order}')
```

#### 상속
- 기존 클래스(부모/상위)의 멤버변수와 메서드를 새로운 클래스(자식/하위)가 그대로 물려받는 것
- 코드 재사용성, 확장

```python
class Parent :
    def __init__(self) :
        self.value = "테스트"
        print("Parent 클래스의 __init__() 호출")

    def test(self) :
        print("Parent 클래스의 test() 호출")

class Child(Parent) :
    pass

child = Child()
child.test()
print(child.value)
```

```python
# super() : 상위 클래스


class Child(Parent) :
    def __init__(self):
        super().__init__()
        print("Child 클래스의 __init__ 호출")

child = Child()
child.test()
print(child.value)
```

```python
class Animal :
    def __init__(self, name) :
        self.name = name
    
    def eat(self) :
        print(f'{self.name}가 먹이를 먹는다')

    def sound(self) :
        print(f'{self.name}이 소리를 낸다.')

class Dog(Animal) :
    def __init__(self, name, kind):
        super().__init__(name)
        self.kind = kind
    
    def fetch(self) :
        print(f'{self.name}이 공을 물어옵니다.')

dog = Dog("초코", "푸들")
dog.eat()
dog.sound()
dog.fetch()
```

```python
# Audio : 멤버변수 : power, volume / 메소드 : 전원 조절, 볼륨조절 
# TV : 멤버변수 : power, volume, size / 메소드 : 전원 조절, 볼륨조절
class AudioVisual :
    def __init__(self, power, volume) :
        self.power = power
        self.volume = volume

    def switch(self, on_off) :
        self.power = on_off

    def setVolume(self, volume) :
        self.volume = volume

    

class Audio(AudioVisual) :
    def __init__(self, power, volume):
        super().__init__(power, volume)

    def tune(self) :
        print("LA la la..." if self.power else "turn it on")

class TV(AudioVisual) : 
    def __init__(self,power, volume, size):
        super().__init__(power, volume)
        self.size = size

    def watch(self) :
        print("hava fun!!" if self.power else "switch on")

tv = TV(False, 12, 50)
# tv 켜기
tv.switch(True)
# tv 보기
tv.watch()

audio = Audio(True, 15)
audio.setVolume(10)
audio.tune()
audio.switch(False)
```

#### 오버라이딩
- 부모 클래스에 있는 메서드를 자식 클래스에서 같은 이름으로 다시 정의 해서 동작 내용 변경

```python
class Animal :
    def __init__(self, name) :
        self.name = name

    def sound(self) :
        print(f'{self.name}가 소리를 냅니다.')

class Dog(Animal) :
    def sound(self) :
        print(f'{self.name} : 멍멍!!')
        return super().sound()
    
class Cat(Animal) :
    def sound(self) :
        # 오버라이딩 시 동일한 부모 메소드를 호출하고 싶다면
        print(f'{self.name} : 야옹!!')
        return super().sound()

dog = Dog("초코")
cat = Cat("몰리")    

dog.sound()
cat.sound()
```

```python
# Employee 부모 클래스 : 
# 멤버 변수 : 이름(name), 급여(pay)
# 멤버 메소드 : showInfo() => 이름, 급여 출력

# Manager 자식 클래스
# 멤버 변수 : 보너스(bonus)
#멤버 메소드 : getTotalPay() : 급여 + 보너스 출력

class Employee :
    def __init__(self, name, pay):
        self.name = name
        self.pay = pay

    def showInfo(self) :
        print(f'이름 : {self.name}, 급여 : {self.pay}')

class Manager(Employee) :
    def __init__(self, name, pay, bonus):
        super().__init__(name, pay)
        self.bonus = bonus

    def getTotalPay(self) :
        print(f'이름 : {self.name}, 급여 : {self.pay} 총 급여 : {self.pay + self.bonus}')

emp = Employee("홍길동", 300)
emp.showInfo()

man = Manager("성춘향", 400, 200)
man.getTotalPay()
```

```python
# Shape 부모 클래스
# 멤버 메서드 : area() => "도형의 넓이" 출력

# Rectangle 자식 클래스
# 멤버 변수 : width, height
# 멤버 메서드 : area() => 직사각형의 넓이 출력

# Circle 자식 클래스
# 멤버 변수 : radius
# 멤버 메서드 : area() => 원의 넓이 출력

class Shape :
    def __init__(self):
        pass

    def area() :
        print("25")

class Rectangle(Shape) :
    def __init__(self, width, height):
        super().__init__()
        self.width = width
        self.height = height

    def area(self) :
        print(f'직사각형의 넓이 : {self.width * self.height}')

class Circle(Shape) :
    def __init__(self, radius):
        super().__init__()
        self.radius = radius
    
    def area(self) :
        print(f'원의 넓이 : {self.radius * self.radius * 3.14}')

rec = Rectangle(10, 20)
rec.area()
cir = Circle(5)
cir.area()
```
