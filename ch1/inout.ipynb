```python
num = input("숫자 입력")

# type(num)
# 사용자에게 숫자를 입력받은 후 +10 한 결과 출력
# TypeError: can only concatenate str (not "int") to str
# print(num + 10)

# ValueError: invalid literal for int() with base 10: 'a'
# try ~ except
if num.isdigit() :
    print(int(num) + 10)
else : 
    print("숫자를 입력하세요.")
```
