````python
# 주차장 프로그램

# 사용자에게 메뉴 제공
# [1] 주차 | [2] 출차 | [3] 종료

# 사용자 메뉴 선택
# 1
# 자동차 A ~ E
# A 자동차 입차. 주차장 상태 ==> ['A']
# 1
# B 자동차 입차. 주차장 상태 ==> ['A', 'B']
# 2
# B 자동차 출차. 주차장 상태 ==> ['A']
# E 자동차 입차. 주차장 상태 ==> ['A', 'B', 'C', 'D', 'E']
# 1
# 주차장 만차

parking_lot = []
top, car_name = 0, 'A'

while True :
    no = int(input("[1] 주차 | [2] 출차 | [3] 종료"))
    
    if no <= 3 :
        if no == 1 :
            # top >= 5 주차장 만차 => 아뭣도 안함
            
            # A 자동차 입차. 주차장 상태 ==> ['A']
            # parking_lot.append()
            # car_name = chr(ord(car_name)+1)
            # top = top + 1
            if top >= 5 :
                print("주차장 만차")
            else :
                parking_lot.append(car_name)
                print(f"{car_name} 자동차 입차. 주차장 상태 ==> {parking_lot}")
                car_name = chr(ord(car_name)+1)
                top = top + 1
        elif no == 2 :
            if top <= 0 :
                print("출차할 차가 없습니다.")
            else :
                parking_lot.remove(chr(ord(car_name)-1))
                print(f"{chr(ord(car_name)-1)} 자동차 출차. 주차장 상태 ==> {parking_lot}")
                car_name = chr(ord(car_name)-1)
                top = top - 1
        else :
            print("프로그램 종료")
            break 
    else :
        print("번호를 확인해주세요.")
    
````

````python
# 새로운 리스트 생성 : 리스트 컴프리헨션

print([ i for i in range(1,7) ])
print([ i * i for i in range(5) ])
print([ i * i for i in [1,3,7,9,16] ])
print([ [i, i+1, i+2] for i in [1,2,3] ])

# 1 ~ 10 사이의 숫자 중에서 홀수 리스트 생성

odd_list = []
for i in range(1,11) :
    if i % 2 == 1 :
        odd_list.append(i)
print(odd_list)

print( [i for i in range(1,10,2)])

odd_list2 = [i for i in range(1,11) if i % 2 == 1]
print(odd_list2)
````

````python
q_list = ["갑", "을", "병", "정", "경"]

# q_list 에서 정을 제외한 요소를 갖는 새로운 리스트 생성
# for
q2_list = [] 
for i in q_list :
    if i != "정" :
        q2_list.append(i)

print(q2_list)

# 리스트 컴프리헨션
q3_list = [ i for i in q_list if i != "정"] 

print(q3_list)
````

````python
q1 = ['nice', 'study', 'python', 'anaconda', '!']

# 5글자 이상의 단어만 새로운 리스트로 생성
# len()

q1_list = []
for i in q1 :
    if len(i) >= 5 :
        q1_list.append(i)

print(q1_list)

print([i for i in q1 if len(i) >= 5])
````

````python
q2 = ['A', 'b', 'c', 'D', 'e', 'F', 'G', 'h']

# 소문자만 추출 후 새로운 리스트로 생성

"X".islower()
"X".isupper()

q2_list = []
for i in q2 :
    if i.islower() :
        q2_list.append(i)

print(q2_list)

print([i for i in q2 if i.islower()])
````
